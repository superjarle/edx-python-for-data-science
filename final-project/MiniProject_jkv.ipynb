We therefore need to normalize the data in order to get meaningful graphs. </br>

## Creating a normalization function
I am therefore creating a function to normalize the input data to be scaled within the range [0,1] </br>
The function uses the column "Value" to be normalized, where the minimum takes the value 0, and the max the value 1. </br>


```python
def norm(a, min, max):
    norm_features = []
    for record in a:
        feature_normed = (record - min) / (max - min)
        norm_features.append(feature_normed)
    return norm_features
```

```python
world_urb_min = world_pop_urban['Value'].min()
world_urb_max = world_pop_urban['Value'].max()
world_gdp_min = world_gdp['Value'].min()
world_gdp_max = world_gdp['Value'].max()
print('The change in World urban population from 1960 to 2015 is %s ' % (world_urb_max - world_urb_min))
```

```python
### Normalizing the Population using the function defined 
normed_pop = norm(world_pop_urban['Value'], world_urb_min, world_urb_max)
normed_gdp = norm(world_gdp['Value'], world_gdp_min, world_gdp_max)
```

**Plot of the normalized data**

```python
plt.figure(figsize=(8, 6))
plt.plot(world_pop_urban['Year'], normed_pop, color= 'Blue')
plt.plot(world_gdp['Year'], normed_gdp, color = 'Green')
plt.xlabel('Year', fontsize = 20)
plt.ylabel('Urban Pop. vs. GDP. (Normalized)', fontsize = 18)
plt.legend(['Urban Population', 'GDP'])
plt.title('Development of Urban population and GDP in the World. 1960 - 2015', fontsize = 20)
plt.show()
```

Secondly, I will examine the development of emissions per Ideally we will see emissions reduce per dollar of GDP over time - this would be an indicator of emissions per capita. ## Investigating economic growth and CO2-emissions

```python
#CO2 emissions (kg per 2005 US$ of GDP)
co2_gdp_data = df[df['IndicatorName'] == 'CO2 emissions (metric tons per capita)']
oecd_co2_cap = co2_gdp_data[co2_gdp_data['CountryName'] == 'OECD members']
latin_co2_cap = co2_gdp_data[co2_gdp_data['CountryName'] == 'Latin America & Caribbean (all income levels)']
oecd_co2_cap.head()
```

```python
latin_co2_cap.head()
```

The figure above does not provide us with maybe as a encouraging narrative as some might have hoped. 

We do observe that the emissions in the OECD countries were highest between 1970 - 1980 per capita, and have been declining since the mid 2000s, but are still a lot higher than for the Latin American countries.

```python
##Finding the difference
oecd_emi = oecd_co2_cap['Value'].max()
lat_emi = latin_co2_cap['Value'].max()

print('The highest difference between OECD and Latin American Emissions in terms of per capita is (in tons!) %s' % (oecd_emi - lat_emi))
```
